```python
#A2

query = """
SELECT DISTINCT
  Nome_Piloto, Nacionalidade_Piloto
FROM
  Piloto NATURAL JOIN Resultado
WHERE
  Valor_Pos_Largada = 1 AND Valor_Pos_Termino = 1 AND Valor_Posicao_Vmr = 1
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#A1

query = """
SELECT DISTINCT
  Nome_Construtor
FROM
  Construtor NATURAL JOIN Resultado NATURAL JOIN Status
WHERE
  Desc_Status = "Velas" OR Desc_Status = "Turbo"
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#A9

query = """
SELECT DISTINCT
  Ano_Corrida, Nome_Construtor, Qte_Pontos
FROM
  Construtor NATURAL JOIN Resultado_Construtor NATURAL JOIN Corrida
WHERE
  Qte_Pontos > 10 AND Id_Circuito = 10 AND Ano_Corrida >= 2016
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#A3

query = """
SELECT DISTINCT
  Nome_Pais, Nome_Circuito
FROM
  Pais NATURAL LEFT JOIN Circuito
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#B2

query = """
SELECT
  Nome_Construtor, DURACAO
FROM(SELECT
      Id_Construtor, Id_Corrida, Id_Piloto, MIN(Duracao_Pitsop) as DURACAO
     FROM
      Pitstop NATURAL JOIN Resultado) NATURAL JOIN Construtor

"""

df = pd.read_sql_query(query, conn)
df
```

```python
#B3

query = """
SELECT
  Nome_Piloto
FROM
  Piloto
WHERE Nacionalidade_Piloto = "Brasileira" AND
  Nome_Piloto NOT IN(
    SELECT DISTINCT
      Nome_Piloto
    FROM
      Piloto NATURAL JOIN Resultado
    WHERE
      Nacionalidade_Piloto = "Brasileira" AND Valor_Pos_Termino <=3
  )
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#B4

query = """
SELECT
  Nacionalidade_Construtor, COUNT(Nacionalidade_Construtor) as ABANDONOS
FROM
  Construtor NATURAL JOIN Resultado NATURAL JOIN Status
WHERE
  Desc_Status = "Embreagem"
GROUP BY
  Nacionalidade_Construtor
HAVING
  ABANDONOS >= 10
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#B9

query = """
SELECT
  Nome_Corrida, Count(Nome_Corrida) as VITORIAS
FROM
  Piloto NATURAL JOIN Resultado  NATURAL JOIN Corrida
WHERE
  Nome_Piloto = "Lewis Hamilton" AND Valor_Pos_Termino = 1
GROUP BY
  Nome_Corrida
"""

df = pd.read_sql_query(query, conn)
df
```

```python
#B5

query = """
SELECT DISTINCT
  Nome_Pais
FROM(
    SELECT
      Nome_Pais, Ano_Corrida, COUNT(*) as NCORRIDAS
    FROM
      Pais NATURAL JOIN Circuito NATURAL JOIN Corrida
    GROUP BY
      Nome_Pais, Ano_Corrida
    HAVING
      NCORRIDAS > 1)
"""

df = pd.read_sql_query(query, conn)
df
```
